### Stair Case Problem - Number of unique ways to reach nth stair 

1. number of ways to reach level 4 is equal to sum of ways to reach level 3 and level 2 -- as we can take 1 or 2 steps
2. this can be recursively solved 
3. similar to fibonacci series, here w(1) - 1 (which is equal to fibonacci(1)), w(2) - 2 (fibonacci(3)) 
4. so w(n) = fibonacci(n+1)

```python
# number of ways to reach step n --- find fibonacci number of n+1

def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def paths(N):
    return fibonacci(N+1)  
N = 3
path = paths(N)
print(path)
```
